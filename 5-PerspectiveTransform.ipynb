```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
image_list = []
title_list = []
for i in range(1,5):
    image_list.append('./data_perspectiveDistortion/dsu' + str(i) + '.jpg')
    title_list.append(['DSU ' + str(i) + ' : Before','DSU ' + str(i) + ' : After Perspective Transform'])
```

```python
img = cv2.imread(image_list[0])
original_pts = np.float32([[322, 279], [759, 311], [325, 535], [766, 508]])
desired_pts = np.float32([[322, 279], [759, 279], [322, 535], [759, 535]])

def PerspectiveTransform(image, original_pts, desired_pts, title_list):
    for i in range(4):
        cv2.circle(image, tuple(original_pts[i]), 5, (0, 0, 255), -1)
    
    matrix = cv2.getPerspectiveTransform(original_pts, desired_pts)
    result = cv2.warpPerspective(image, matrix, (1024, 768))

    res_show = []
    res_show.append(image)
    res_show.append(result)
    
    fig = plt.figure(figsize=(16, 9))

    for i in range(2):
        fig.add_subplot(1,2,i+1)
        plt.title(title_list[i])
        plt.imshow(cv2.cvtColor(res_show[i], cv2.COLOR_BGR2RGB))
        plt.axis('off')

    plt.tight_layout()
    plt.show()


PerspectiveTransform(img, original_pts, desired_pts, title_list[0])
```

```python
def run():
    img = cv2.imread(image_list[1])
    original_pts = np.float32([[180, 237], [822, 219], [178, 503], [821, 514]])
    desired_pts = np.float32([[180, 237], [822, 237], [180, 503], [822, 503]])
    PerspectiveTransform(img, original_pts, desired_pts, title_list[1])
    
```
