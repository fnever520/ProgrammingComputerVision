```python
import numpy as np
import cv2
from PIL import Image
import os
import matplotlib.pyplot as plt

input_image = "data/book_perspective.jpg"
img = cv2.imread(input_image)
#img = np.array(img, dtype=np.uint8)
real_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(16,9))
plt.imshow(gray)
plt.show()
```

```python

contours = cv2.Canny(gray, 60,200)
thresh = cv2.adaptiveThreshold(contours, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15,1)

plt.figure(figsize=(16,9))
plt.title("The edge detection")
plt.imshow(contours, cmap='gray')
plt.show()
```

```python
'''
kernel = np.array([[-1,-1,-1],
                   [-1, 9,-1],
                   [-1,-1,-1]])
                   
'''

kernel_sharpened = np.array([[0.,-1.,0.],
                           [-1.,5.,-1.],
                           [0.,-1.,0.]])

kernel_blur = np.ones((3,3), np.float32)/9

kernel_gaussianBlur = np.array([[1.,2.,1.],
                                [2.,4.,2.],
                                [1.,2.,1.]])/16


sharpened = cv2.filter2D(real_image,-1, kernel_sharpened)
plt.figure(figsize=(16,9))
plt.title("Sharpened")
plt.imshow(sharpened)
plt.show()
```

```python
blurring = cv2.filter2D(real_image, -1, kernel_blur)
plt.figure(figsize=(16,9))
plt.title("Blurring")
plt.imshow(blurring)
plt.show()
```

```python
gaussianblur = cv2.filter2D(real_image, -1, kernel_gaussianBlur)
plt.figure(figsize=(16,9))
plt.title("Gaussian Blur")
plt.imshow(gaussianblur)
plt.show()
```
